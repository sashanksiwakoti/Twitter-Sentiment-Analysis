```python
import pandas as pd
import warnings
import seaborn as sns
import re
%matplotlib inline
warnings.filterwarnings('ignore') 
```

```python
#fetching csv file from desktop
df=pd.read_csv("twitter Sentiments.csv")
df.head()
```

```python
#pattern removing from the input text
def RemPattern(inp_text, pattern):
    rgx=re.findall(pattern, inp_text)
    for wrd in rgx:
        inp_text=re.sub(wrd,'',inp_text)
    return inp_text
```

```python
df.head()
```

```python
# removing @user twitter handles
#uses numpy library class vectorize for dataframing
import numpy as np 
df['clean twitter']=np.vectorize(RemPattern)(df['tweet'],"@[\w]*")
```

```python
df.head()
```

```python
#removing punctuation, chars, and nums
df['clean twitter']=df['clean twitter'].str.replace("[^a-zA-Z#]", " ")
df.head()
```

```python
#taking out little words
df['clean twitter']=df['clean twitter'].apply(lambda z: " ".join([w for w in z.split() if len(w) > 3]))
#from the new dataframe(clean twitter)
df.head()
```

```python
#making the words single and making it into a token
tokeni= df['clean twitter'].apply(lambda z: z.split())
tokeni.head()
```

```python
#using stem frm nltk library(porterstemmer) for stemming
import nltk
from nltk.stem.porter import PorterStemmer
stem_fun = PorterStemmer()
tokeni=tokeni.apply(lambda phrase: [stem_fun.stem(word) for word in phrase])
tokeni.head()
```

```python
#merging words in a single sentence
for j in range(len(tokeni)):
    tokeni[j]=" ".join(tokeni[j])
df['clean twitter']=tokeni
df.head()
```

```python
#extrawithdrawct the '#'
def hash_tags_withdrawal(tweets):
    hashT=[]
    # looping words in tweets
    for twt in tweets:
        ht=re.findall(r"#(\w+)", twt)
        hashT.append(ht)
    return hashT
```

```python
#withdraw tweets from positive tweets
Ht_Pos=hash_tags_withdrawal(df['clean twitter'][df['label']==0])
#withdraw tweets from negative tweets
Ht_Neg=hash_tags_withdrawal(df['clean twitter'][df['label']==1])
```

```python
Ht_Pos[:4]
```

```python
#unnesting the list
Ht_Pos=sum(Ht_Pos, [])
Ht_Neg=sum(Ht_Neg, [])
```

```python
Ht_Pos[:4]
```

```python
freq=nltk.FreqDist(Ht_Pos)
P = pd.DataFrame({'Hashtags': list(freq.keys()), 'Count': list(freq.values())})
P.head()
```

```python
# swlwcting the top 10 '#' for +ive
P=P.nlargest(columns='Count', n =10)
plt.figure(figsize=(14,8))
sns.barplot(data=P, x = 'Hashtags', y='Count')
plt.show()
```

```python
# for -tive
freq=nltk.FreqDist(Ht_Neg)
P = pd.DataFrame({'Hashtags': list(freq.keys()), 'Count': list(freq.values())})
P.head()
```

```python
# swlwcting the top 10 '#' for +ive
P=P.nlargest(columns='Count', n =10)
plt.figure(figsize=(15,9))
sns.barplot(data=P, x = 'Hashtags', y='Count')
plt.show()
```

```python
#feature extraction
from sklearn.feature_extraction.text import CountVectorizer
bow_vectorizer = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')
bow= bow_vectorizer.fit_transform(df['clean twitter'])
```

```python
#bow[0].toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(bow,df['label'], random_state=42, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
#training
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Testing
pred= model.predict(x_test)
f1_score(y_test, pred)
```

```python
accuracy_score(y_test,pred)
```
